#### sleep_in

```python
def sleep_in(weekday, vacation):
  if not weekday or vacation:
    return True
  else:
    return False

print(sleep_in(True, True)) # do you sleep in on a weekday thats a holiday?
print(sleep_in(True, False)) # do ou sleep in on a weekday thats not a holiday?
print(sleep_in(False, True)) # do you sleep in on a weekend that's a holiday?
print(sleep_in(False, False)) # do ou sleep in on a weekend thats not a holiday?

```

#### diff21


```python
def diff21(n):
  x = 21 # the number subtracted from is always 21
  if n > 21: #if the number is > 21
    return 2*(n-x) # two times the difference
  else :
    return x - n # else just the diff
```

```python
print(diff21(8)) # 21 minus 8
print(diff21(28)) # two times 7
```

#### missing_char

```python
def missing_char(str, n):
  a = str[:n] # get everything up to that index or numbered element
  b = str[n+1:] # get everything after that numbered index
  return a+b # paste them together


print(missing_char('kitten', 1)) # remove the second element
print(missing_char('kitten', 0)) 
print(missing_char('kitten', 4)) 
```

#### monkey trouble

```python
def monkey_trouble(a_smile, b_smile):
# if a is smiling and b not or b is smiling and a not
  if (a_smile and not b_smile) or (not a_smile and b_smile) :
    return False
  else:
    return True

print(monkey_trouble(True, True))
print(monkey_trouble(False, False))
print(monkey_trouble(True, False))
```

### parrot trouble

```python
def parrot_trouble(talking, hour):
  if talking and (hour < 7 or hour > 20):
    return True
  else:
    return False

print(parrot_trouble(True, 6)) 
print(parrot_trouble(True, 7)) 
print(parrot_trouble(False, 6)) 
```

#### positive negative

```python
def pos_neg(a, b, negative):
  if negative:
      if (a<0 and b <0):
          return True
      else:
          return False
  else:
      if (a>0 and b>0) or (b<0 and a<0) :
        return False
      else:
        return True

print(pos_neg(1, -1, False))
print(pos_neg(-1, 1, False))
print(pos_neg(-4, -5, True))
```

#### front back

```python
def front_back(str):
    if len(str)>1:
        first = str[0]
        last = str[-1]
        return last+str[1:-1] + first
    else:
        return str

print(front_back('code')) 
print(front_back('a')) 
print(front_back('ab'))
```

#### print double

```python
def sum_double(a, b):
  if a==b:
    return 2 * (a+b)
  else:
    return a+b

print(sum_double(1,2))
print(sum_double(3,2))
print(sum_double(2,2))
```

#### makes 10

```python
def makes10(a, b):
  list = [a,b]
  if 10 in list:
    return True
  elif a+b == 10:
    return True
  else:
    return False

print(makes10(9,10))
print(makes10(9,9))
print(makes10(1,9))
```

#### not_string

```python
def not_string(str):
  if str[:3]=="not":
    return str
  else:
    return "not " + str

print(not_string('candy')) 
print(not_string('x'))
print(not_string('not bad'))
```

#### front3

```python
def front3(str):
    if len(str)<3:
        return str*3
    else:
        return str[:3]*3

print(front3('Java'))
print(front3('Chocolate'))
print(front3('abc'))
```

#### string times

```python
def string_times(str, n):
  if n > 0:
    return str*n
  else:
    return ''

print(string_times('Hi', 2))
print(string_times('Hi', 3))
print(string_times('Hi', 1))
```

#### string splosion

```python
def exp_word(str):
    holder=''
    for i in range(len(str)+1):
        holder+=str[:i]
    return holder

print(exp_word('Code'))
print(exp_word('abc'))
print(exp_word('apple'))
```

#### last 2

```python
def last2(str):
    oth = str[:-1] # I want to include everything up until the last letter
    last=str[-2:] # get the last two letters as a matching element
    tots = 0
    for i in range(len(oth)): #for every latter in the oth element
        check = oth[i:i+2] # get that letter through the next one
        if check == last: # check it against the last 2
            tots+=1 # add 1 if the check elemtn matches
    return tots # if non match you get 0
```

#### array123

```python
def array123(nums, checks=[1,2,3]):
    for i in range(len(nums)): # for every element in nums
        a = nums[i:i+3] # get 3 number combinations
        if a == checks: # if that 3 element slice = checks
            return True # return true - once one matches this stops
    return False # this needs to be outside the loop or the loop stops on the first iteration 

print(array123([1, 1, 2, 3, 1]))# → True
print(array123([1, 1, 2, 4, 1])) # → False
print(array123([1, 1, 2, 1, 2, 3]))# → True
```
