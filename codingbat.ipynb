#### sleep_in

```python
def sleep_in(weekday, vacation):
  if not weekday or vacation:
    return True
  else:
    return False

print(sleep_in(True, True)) # do you sleep in on a weekday thats a holiday?
print(sleep_in(True, False)) # do ou sleep in on a weekday thats not a holiday?
print(sleep_in(False, True)) # do you sleep in on a weekend that's a holiday?
print(sleep_in(False, False)) # do ou sleep in on a weekend thats not a holiday?

```

#### diff21


```python
def diff21(n):
  x = 21 # the number subtracted from is always 21
  if n > 21: #if the number is > 21
    return 2*(n-x) # two times the difference
  else :
    return x - n # else just the diff
```

```python
print(diff21(8)) # 21 minus 8
print(diff21(28)) # two times 7
```

```python
def near_hundred(n):
  a = 100 
  b = 200
  # if the abs difference between the number and 100 or 200 is less than 11 return true
  if abs(n-a) <=10 or abs(n-b)<=10: # 
    return True
  else:
    return False


print(near_hundred(80))
print(near_hundred(98))
print(near_hundred(108))
print(near_hundred(198))
print(near_hundred(208))
print(near_hundred(700))
```

#### missing_char

```python
def missing_char(str, n):
  a = str[:n] # get everything up to that index or numbered element
  b = str[n+1:] # get everything after that numbered index
  return a+b # paste them together


print(missing_char('kitten', 1)) # remove the second element
print(missing_char('kitten', 0)) 
print(missing_char('kitten', 4)) 
```

#### monkey trouble

```python
def monkey_trouble(a_smile, b_smile):
# if a is smiling and b not or b is smiling and a not
  if (a_smile and not b_smile) or (not a_smile and b_smile) :
    return False
  else:
    return True

print(monkey_trouble(True, True))
print(monkey_trouble(False, False))
print(monkey_trouble(True, False))
```

### parrot trouble

```python
def parrot_trouble(talking, hour):
  if talking and (hour < 7 or hour > 20):
    return True
  else:
    return False

print(parrot_trouble(True, 6)) 
print(parrot_trouble(True, 7)) 
print(parrot_trouble(False, 6)) 
```
